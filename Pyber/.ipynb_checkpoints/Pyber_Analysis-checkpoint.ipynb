```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to load 
city_data_file = "city_data.csv"
ride_data_file = "ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_file)
ride_data = pd.read_csv(ride_data_file)

# Combine the data into single dataset
full_data = pd.merge(ride_data, city_data, on="city")

# Display the data table for preview
full_data.head()
```

## Bubble Plot of Ride Sharing Data 

```python
# Obtain the x and y coordinates for each of the three city types 
city_types = ["Urban", "Suburban", "Rural"]
city_types_avg_fare = [pd.DataFrame(full_data.loc[full_data["type"] == city_type].groupby(["city", "driver_count"])["fare"].mean()).reset_index() for city_type in city_types]
city_types_total_ride = [full_data.loc[full_data["type"] == city_type].groupby("city")["ride_id"].count() for city_type in city_types]

# Build the scatter plots for each city types
colors = ["coral", "lightskyblue", "gold"]
[plt.scatter(city_types_total_ride[item], 
             city_types_avg_fare[item]["fare"], 
             c=colors[item],
             s=10*city_types_avg_fare[item]["driver_count"], 
             edgecolors="black", alpha=0.75) for item in np.arange(3)]

# Incorporate the other graph properties
plt.grid()
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare (Per City)")
plt.title("Pyber Ride Sharing Data (2016)")

# Create a legend
plt.legend(title="City Types", labels=city_types)
plt.text(45, 35,"Note: Circle size correlates with driver count per city.")

# Save Figure
plt.savefig("./pyber_ride_sharing_data.png")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate total fare by Type 
city_types = ["Urban", "Suburban", "Rural"]
city_types_total_fare = [full_data.loc[full_data["type"] == city_type]["fare"].sum() for city_type in city_types]

# Build Pie Chart
colors = ["lightcoral", "lightskyblue", "gold"]
plt.pie(city_types_total_fare, explode=[0.1, 0, 0], 
        labels=city_types, colors=colors, shadow=True, 
        autopct="%.1f%%", startangle=280)

plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("./total_fares_by_city_type.png")

# Show plot
plt.show()
```

## Total Rides by City Type

```python
# Calculate total ride by Type 
city_types = ["Urban", "Suburban", "Rural"]
city_types_total_ride = [full_data.loc[full_data["type"] == city_type]["ride_id"].count() for city_type in city_types]

# Build Pie Chart
colors = ["lightcoral", "lightskyblue", "gold"]
plt.pie(city_types_total_ride, explode=[0.1, 0, 0], 
        labels=city_types, colors=colors, shadow=True, 
        autopct="%.1f%%", startangle=280)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("./total_rides_by_city_type.png")

# Show plot
plt.show()
```

## Total Drivers by City Type

```python
# Calculate total drivers by Type 
city_types = ["Urban", "Suburban", "Rural"]
city_types_total_driver = [pd.DataFrame(full_data.loc[full_data["type"] == city_type].groupby(["city", "driver_count"]).count()).reset_index()["driver_count"].sum() for city_type in city_types]

# Build Pie Chart
colors = ["lightcoral", "lightskyblue", "gold"]
plt.pie(city_types_total_driver, explode=[0.1, 0, 0], 
        labels=city_types, colors=colors, shadow=True, 
        autopct="%.1f%%", startangle=240)
plt.title("% of Total Driver by City Type")

# Save Figure
plt.savefig("./total_drivers_by_city_type.png")

# Show plot
plt.show()
```

## Observable trends based on data.

- Rural has a small number of rides but a high average fare. The lack of rides makes rural occupy the smallest share in total fare.
- Suburban has more number of rides comparied to rural, and a higher average fare comparied to urban. 
- Urban has the largest number of rides and the lowest average fare. The largest number of rides and drivers, make urban take 62.7% share of total fare.
